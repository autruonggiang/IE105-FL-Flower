```python
import os
import subprocess
import shutil
from tqdm.auto import tqdm

def create_image_from_exe(input_path, output_path):
    output_path = '~/NLP/DAINB/' + output_path
    input_path = '~/NLP/DAINB/' + input_path
    # Tạo ảnh từ file exe
    command = f"java -jar ~/NLP/DAINB/PortEx/progs/PortexAnalyzer.jar -p {output_path} {input_path}"
    subprocess.run(command, shell=True, stdout=subprocess.DEVNULL)

def process_folder(folder_path):
    # Lặp qua các file trong thư mục
    root_path = 'DikeDataset/files'

    files_in_folder_a = [i + '.png' for i in os.listdir(os.path.join(root_path, folder_path))]

    # Get a list of files in folder B
    files_in_folder_b = os.listdir(os.path.join('Dataset', folder_path))
    # Identify files in folder A that are not in folder B
    files_to_delete = set(files_in_folder_b) - set(files_in_folder_a)
    print('REMOVE: ', len(files_to_delete))
    # Delete files in folder A that are not in folder B
    for file_name in files_to_delete:
        file_path = os.path.join('Dataset', folder_path, file_name)
        try:
            os.remove(file_path)
        except:
            shutil.rmtree(file_path)

    for filename in tqdm(os.listdir(os.path.join(root_path, folder_path)), desc = folder_path):
        file_path = os.path.join(root_path, folder_path, filename)
        # Kiểm tra nếu là file exe
        if filename.endswith(".exe"):
            # Tạo đường dẫn cho file ảnh đầu ra
            output_image_path = os.path.join('Dataset', folder_path, f"{filename}.png")
            if f"{filename}.png" in os.listdir(os.path.join('Dataset', folder_path)):
                continue

            # Tạo ảnh từ file exe
            create_image_from_exe(file_path, output_image_path)

# # Thư mục chứa các file benign
benign_folder = "benign"
process_folder(benign_folder)

# Thư mục chứa các file malware
malware_folder = "malware"
process_folder(malware_folder)
```
